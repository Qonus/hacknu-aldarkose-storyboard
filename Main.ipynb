```python
from openai import OpenAI
import json

client = OpenAI(api_key="") 

def ask(prompt):
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini-2024-07-18",
            messages=[
                {"role": "system", "content": '''
                You are a helpful assistant designed to output story boards about Aldar Kose in JSON.
                Aldar Kose is a character from a 3d animated cartoon series written after:
                Aldar kose, (native name Kazakh: Алдар көсе/Aldar köse) is a Kazakh folk fairy tale and name of main character.[1] He is the collective image of the sly but very kind man. In fairy tales he is a swindler, cheating the greedy rich, evil Khans and helping the poor and weak people.[2][3] The Historical Dictionary of Kazakhstan refers to him as a "creative liar".
                User gives you a prompt for a story.
                You take it and generate 6-10 story cards in "storyboard". Each card has to include title and a description of it in: "content" and a prompt for an image generator: "scene".
                Your output is strictly JSON, story board format.
                The scene should be very descriptive including clothes, colors, and many things must be repeated for generative model because it doesnt have the context, it just has it's prompt.
                Be more visual in your descriptions. Describe camera angles, shots, lightning, characters, environment, clothes colors and everything
                Also everytime describe the setting of these 3d cartoon series, it means you should say how it looks in the series.
                Feel free to make scene description as long as you like.
                '''},
                {"role": "user", "content": prompt}
            ],
            response_format={"type": "json_object"} # Crucial parameter for JSON output
        )
    
        # The model's response is a string that contains the JSON object
        json_string = response.choices[0].message.content
        
        # Convert the JSON string to a Python dictionary
        python_dict = json.loads(json_string)
        return python_dict
        
    except Exception as e:
        print(f"An error occurred during the API call: {e}")
```

```python
import io, sqlite3, datetime, random    
from PIL import Image  
import torch
from diffusers import FluxPipeline
from pathlib import Path
import matplotlib.pyplot as plt
from typing import List, Tuple 

BASE_MODEL_ID = "black-forest-labs/FLUX.1-dev"
LOCAL_LORA_PATH = "./workspace/ai-toolkit/output/aldar.safetensors" 
LORA_SCALE = 0.7
pipe = FluxPipeline.from_pretrained(BASE_MODEL_ID, torch_dtype=torch.bfloat16)
pipe.enable_model_cpu_offload()

pipe.load_lora_weights(
    Path(LOCAL_LORA_PATH).parent, 
    weight_name=Path(LOCAL_LORA_PATH).name
)

prompt = "Aldar Kose out grew his taqiyah and he decided to not wear it ever again"
prompt = "Aldar Kose with horses jumping on water"
response = ask(prompt)['storyboard']
print(response)
DB_PATH = "storyboards.db"

def db_connect(db_path: str = DB_PATH):
    conn = sqlite3.connect(db_path)
    conn.execute("""
        CREATE TABLE IF NOT EXISTS storyboard (
          seed INTEGER PRIMARY KEY,
          prompt TEXT NOT NULL,
          created_at TEXT NOT NULL
        )
    """)
    conn.execute("""
        CREATE TABLE IF NOT EXISTS storyboard_image (
          id INTEGER PRIMARY KEY AUTOINCREMENT,
          seed INTEGER NOT NULL,
          idx INTEGER NOT NULL,
          caption TEXT,
          scene TEXT,
          image_png BLOB NOT NULL,
          FOREIGN KEY(seed) REFERENCES storyboard(seed)
        )
    """)
    return conn

def pil_to_bytes(img: Image.Image) -> bytes:
    buf = io.BytesIO()
    img.save(buf, format="PNG")
    return buf.getvalue()

def save_storyboard(seed: int, prompt: str, response: list, images: List[Image.Image], db_path: str = DB_PATH):
    conn = db_connect(db_path)        
    with conn:
        conn.execute(
            "INSERT OR REPLACE INTO storyboard(seed, prompt, created_at) VALUES (?, ?, ?)",
            (seed, prompt, datetime.datetime.utcnow().isoformat())
        )
        for i, (frame, img) in enumerate(zip(response, images)):
            conn.execute(
                "INSERT INTO storyboard_image(seed, idx, caption, scene, image_png) VALUES (?, ?, ?, ?, ?)",
                (seed, i, frame.get('caption', ''), frame.get('scene', ''), pil_to_bytes(img))
            )
    conn.close()

def load_storyboard(seed: int, db_path: str = DB_PATH) -> List[Tuple[int, str, str, Image.Image]]:
    conn = db_connect(db_path)
    cur = conn.execute(
        "SELECT idx, caption, scene, image_png FROM storyboard_image WHERE seed=? ORDER BY idx ASC",
        (seed,)
    )
    rows = cur.fetchall()
    conn.close()
    out = []
    for idx, caption, scene, blob in rows:
        img = Image.open(io.BytesIO(blob)).convert("RGBA")
        out.append((idx, caption, scene, img))
    return out


master_seed = random.getrandbits(63)
images = []

for i in range(len(response)):
    print(response[i]['content'])
    image = pipe(
        "Only one Aldar Kose. Villagers look different than Aldar. In Poor 3d animation style\n" + response[i]['scene'] + "\nOnly one Aldar Kose, no clones, no twin, no duplicate, no clone, no reflection.",
        height=512,
        width=512,
        guidance_scale=3.5,
        num_inference_steps=25,
        # lora_scale=LORA_SCALE,
        max_sequence_length=512,
        generator=torch.Generator("cpu").manual_seed(master_seed+i)
    ).images[0]
    images.append(image)
    # image.save(f"flux-dev{i}.png")
    fig, ax = plt.subplots()
    save_storyboard(master_seed, prompt, response, images)
    ax.axis("off")
    ax.imshow(image)
    plt.show()

print(master_seed)
```

```python
import cv2
import numpy as np
from tqdm import tqdm

def pil_to_bgr(img_pil: Image.Image) -> np.ndarray:
    if img_pil.mode in ("RGBA", "LA"):
        bg = Image.new("RGB", img_pil.size, (255, 255, 255))
        bg.paste(img_pil, mask=img_pil.split()[-1])
        img_pil = bg
    else:
        img_pil = img_pil.convert("RGB")
    arr = np.array(img_pil)[:, :, ::-1] 
    return arr.astype(np.uint8)

def to_uint8(img: np.ndarray) -> np.ndarray:
    img = np.clip(img, 0, 255)
    return img.astype(np.uint8)

def resize_to_common(a: np.ndarray, b: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    ha, wa = a.shape[:2]
    hb, wb = b.shape[:2]
    if (ha, wa) != (hb, wb):
        b = cv2.resize(b, (wa, ha), interpolation=cv2.INTER_CUBIC)
    return a, b

def compute_flow(a_bgr: np.ndarray, b_bgr: np.ndarray, backend: str = "DIS") -> np.ndarray:
    a_gray = cv2.cvtColor(a_bgr, cv2.COLOR_BGR2GRAY)
    b_gray = cv2.cvtColor(b_bgr, cv2.COLOR_BGR2GRAY)

    if backend.upper() == "DIS" and hasattr(cv2, "DISOpticalFlow_create"):
        dis = cv2.DISOpticalFlow_create(cv2.DISOPTICAL_FLOW_PRESET_ULTRAFAST)
        flow = dis.calc(a_gray, b_gray, None)  
    else:
        flow = cv2.calcOpticalFlowFarneback(
            a_gray, b_gray, None,
            pyr_scale=0.5, levels=3, winsize=15,
            iterations=3, poly_n=5, poly_sigma=1.2, flags=0
        )
    return flow

def warp_img(img_bgr: np.ndarray, flow: np.ndarray) -> np.ndarray:
    h, w = img_bgr.shape[:2]
    grid_x, grid_y = np.meshgrid(np.arange(w), np.arange(h))
    map_x = (grid_x + flow[..., 0]).astype(np.float32)
    map_y = (grid_y + flow[..., 1]).astype(np.float32)
    warped = cv2.remap(
        img_bgr, map_x, map_y,
        interpolation=cv2.INTER_LINEAR,
        borderMode=cv2.BORDER_REFLECT101
    )
    return warped

def interpolate_pair(a_bgr: np.ndarray, b_bgr: np.ndarray, k: int = 8, backend: str = "DIS") -> List[np.ndarray]:
    a_bgr, b_bgr = resize_to_common(a_bgr, b_bgr)
    flow_ab = compute_flow(a_bgr, b_bgr, backend=backend)
    flow_ba = compute_flow(b_bgr, a_bgr, backend=backend)

    inter_frames = []
    for i in range(1, k + 1):
        t = i / (k + 1)
        warp_a = warp_img(a_bgr, flow_ab * t)
        warp_b = warp_img(b_bgr, flow_ba * (1.0 - t))
        blend = to_uint8((1.0 - t) * warp_a + t * warp_b)
        inter_frames.append(blend)
    return inter_frames
```

```python
def interpolate_sequence_rife(key_bgr: List[np.ndarray], inter_frames: int,
                              rife_exe: str = RIFE_EXE, rife_model: str = RIFE_MODEL,
 
    if len(key_bgr) < 2:
        raise ValueError("Нужно минимум 2 ключевых кадра для RIFE.")

    from shutil import which
    if which(rife_exe) is None and not Path(rife_exe).exists():
        raise FileNotFoundError(
            f"Не найден '{rife_exe}'. Установите rife-ncnn-vulkan и укажите путь в RIFE_EXE."
        )

    with tempfile.TemporaryDirectory() as tmpdir_in, tempfile.TemporaryDirectory() as tmpdir_out:
        tmpdir_in = Path(tmpdir_in)
        tmpdir_out = Path(tmpdir_out)

        h, w = key_bgr[0].shape[:2]
        for i in range(1, len(key_bgr)):
            if key_bgr[i].shape[:2] != (h, w):
                key_bgr[i] = cv2.resize(key_bgr[i], (w, h), interpolation=cv2.INTER_CUBIC)

        for i, img in enumerate(key_bgr):
            cv2.imwrite(str(tmpdir_in / f"{i:08d}.png"), img)

        M = len(key_bgr)
        desired_total = M + (M - 1) * inter_frames

        cmd = [
            str(rife_exe),
            "-i", str(tmpdir_in),
            "-o", str(tmpdir_out),
            "-n", str(desired_total),      
            "-f", "%08d.png",           
            "-m", str(rife_model)
        ]
        if verbose:
            cmd.insert(1, "-v")

        print(">> RIFE:", " ".join(cmd))
        try:
            subprocess.run(cmd, check=True)
        except subprocess.CalledProcessError as e:
            raise RuntimeError(f"rife-ncnn-vulkan завершился с ошибкой: {e}")

        out_files = sorted(glob.glob(str(tmpdir_out / "*.png")))
        if len(out_files) != desired_total:
            print(f"Предупреждение: ожидалось {desired_total} кадров, найдено {len(out_files)}.")
        timeline = []
        for fp in out_files:
            bgr = cv2.imread(fp, cv2.IMREAD_COLOR)
            if bgr is None:
                continue
            timeline.append(bgr.astype(np.uint8))
        return timeline
```

```python
key_bgr = [pil_to_bgr(img) for _, img in frames]

if FLOW_BACKEND.upper() == "RIFE":
    timeline = interpolate_sequence_rife(
        key_bgr, INTER_FRAMES, rife_exe=RIFE_EXE, rife_model=RIFE_MODEL, verbose=RIFE_VERBOSE
    )
else:
    timeline = []
    for i in range(len(key_bgr) - 1):
        A = key_bgr[i]
        B = key_bgr[i + 1]
        timeline.append(A)
        inter = interpolate_pair(A, B, k=INTER_FRAMES, backend=FLOW_BACKEND)
        timeline.extend(inter)
    timeline.append(key_bgr[-1])
```

```python
h, w = timeline[0].shape[:2]
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
output_fps = FPS * (INTER_FRAMES + 1) 
vw = cv2.VideoWriter(str(OUTPUT_MP4), fourcc, output_fps, (w, h))

for frame in tqdm(timeline, desc="Writing video"):
    vw.write(frame)
vw.release()

print(f"✅ Видео сохранено: {OUTPUT_MP4} | frames={len(timeline)} | fps={output_fps}")
```
